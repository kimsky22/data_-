```python
#@markdown  **üêπ Student Activity 1_2 ‚§µÔ∏è**

#@markdown  **Select rows Î≥µÏàòÌñâ ÏÑ†ÌÉù**

#@markdown  **Row name[] with the colon symbols**

import pandas as pd

exam_data = {'Math':[90, 80, 70, 85], 'English':[98, 89, 95, 79],
             'Music':[85, 95, 100, 100], 'Gym':[100,90,90, 88]}

df = pd.DataFrame(exam_data, index = ['Simon', 'Emma', 'Mary', 'Louis'])

#The first colon : indicates that we're selecting all rows.
#The second colon : indicates the step size, which is -1. A step size of -1 means we're traversing the array in reverse order.
df.iloc[::-1]
```

```python
#@markdown  **üêπ Student Activity 1_3 ‚§µÔ∏è**

#@markdown  **Select rows Î≥µÏàòÌñâ ÏÑ†ÌÉù**

#@markdown  **Row names selected [ ] with the colon symbols**

import pandas as pd

exam_data = {'Math':[90, 80, 70, 85], 'English':[98, 89, 95, 79],
             'Music':[85, 95, 100, 100], 'Gym':[100,90,90, 88]}

df = pd.DataFrame(exam_data, index = ['Simon', 'Emma', 'Mary', 'Louis'])

#You will only get the last code line of >> df.iloc[::-1]
df.iloc[::2]
df.iloc[0:3:2]
df.iloc[::-1]
```

```python
#@markdown  **üêπ Student Activity 1_4 ‚§µÔ∏è**

#@markdown  **Select rows Î≥µÏàòÌñâ ÏÑ†ÌÉù**

#@markdown  **column name[] with the colon symbol**

import pandas as pd

exam_data = {'Math':[90, 80, 70, 85], 'English':[98, 89, 95, 79],
             'Music':[85, 95, 100, 100], 'Gym':[100,90,90, 88]}

df = pd.DataFrame(exam_data, index = ['Simon', 'Emma', 'Mary', 'Louis'])

#How can you get all three tables?
df1 = df.iloc[::2]
df2 = df.iloc[0:3:2]
df3 = df.iloc[::-1]

print(df1)
print('\n')
print(df2)
print('\n')
print(df3)
```
